```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
m0 = 400 # масса ракеты на старте (т)
mp = m0*0.02 # полезная нагрузка (т) - 2% от m0
ms = m0*0.08 # структурная масса (т) - 8% от m0
mt = m0*0.90 # масса топлива (т) - 90% от m0
u = 3 # скорость истечения сгорающего топлива (км/с)
```

```python
mtt = np.arange(mt, 0, -0.0005) # изменение массы топлива в полёте
V = u*np.log(m0/(ms+mp+mtt)) # изменение скорости
Vmax = max(V) # максимально достижимая скорость
V1 = 7.91 # первая космическая скорость (км/с)
```

```python
print("Vmax:", Vmax)
print("V1:", V1)
plt.title("Зависимость скорости V от массы топлива mt") # заголовок
plt.xlabel("mt, т") # ось абсцисс
plt.ylabel("V, км/с") # ось ординат
plt.grid() # включение отображение сетки
a = plt.plot(mtt, V, 'b') # построение графика
plt.gca().invert_xaxis() # реверс оси абсцисс
a = plt.plot(0, Vmax, '*g') # максимально достижимая скорость
a = plt.plot(0, V1, '*r') # первая космическая скорость
```
